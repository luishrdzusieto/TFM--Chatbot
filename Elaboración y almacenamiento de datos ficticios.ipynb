```python
from pymongo import MongoClient
import random

# Conexión a MongoDB Atlas
client = MongoClient("mongodb+srv://frandosramos2:S9nTtOiFoOS8xNRC@cluster0.kw5ss.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0")
db = client['aseguradora']
asegurados = db['asegurados']
coste_reparaciones = db['coste_reparaciones']

# Factores de coste base y complejidad por componente, ajustados por realidad de mercado
coste_bases = {
    'Puerta': (100, 250),
    'Faro': (80, 180),
    'Parabrisas': (150, 300),
    'Luna': (120, 250),
    'Ventana': (80, 160),
    'Techo': (130, 280),
    'Rueda': (70, 150),
    'Parachoques': (100, 220),
    'Guardabarros': (90, 200),
    'Capo': (120, 260),
    'Maletero': (100, 240),
    'Motor': (300, 650)
}

factor_mano_obra = {
    'Puerta': 1.1,
    'Faro': 1.05,
    'Parabrisas': 1.15,
    'Luna': 1.1,
    'Ventana': 1.05,
    'Techo': 1.15,
    'Rueda': 1.0,
    'Parachoques': 1.1,
    'Guardabarros': 1.05,
    'Capo': 1.1,
    'Maletero': 1.1,
    'Motor': 1.3
}

def calcular_coste_material(años_vehiculo, base_min, base_max):
    # Disminución de coste material con la edad del vehículo
    factor_años = 1 - (min(años_vehiculo, 20) / 20) * 0.5  # hasta un 50% menos para vehículos de 20 años o más
    return round(random.uniform(base_min, base_max) * factor_años, 2)

def calcular_coste_mano_obra(coste_material, factor_complexidad):
    return round(coste_material * factor_complexidad, 2)

# Generar costos para cada componente
for asegurado in asegurados.find():
    marca = asegurado['marca_vehiculo']
    modelo = asegurado['modelo_vehiculo']
    años_vehiculo = asegurado['años_vehiculo']

    for componente, (base_min, base_max) in coste_bases.items():
        coste_material = calcular_coste_material(años_vehiculo, base_min, base_max)
        coste_mano_obra = calcular_coste_mano_obra(coste_material, factor_mano_obra[componente])
        
        document = {
            'marca_vehiculo': marca,
            'modelo_vehiculo': modelo,
            'años_vehiculo': años_vehiculo,
            'componente': componente,
            'coste_material': coste_material,
            'coste_mano_obra': coste_mano_obra
        }
        coste_reparaciones.insert_one(document)
```
